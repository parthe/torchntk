```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
CNN = [[59.97, 60.20, 64.11, 69.48, 75.57], [63.81, 80.93, 83.75, 82.92, 83.30]]
CNTK2k = [[40.94,42.54,43.43,43.42,42.53], [49.71,51.06,51.73,51.92,52.22]]
```

```python
from scipy.stats import kendalltau
```

```python
print(kendalltau(CNTK2k[0],CNN[0]))
print(kendalltau(CNTK2k[1],CNN[1]))
```

```python
plt.plot(CNTK2k[0],CNN[0],'k.')
plt.plot(CNTK2k[1],CNN[1],'.',color='tab:blue',)
plt.xlabel('CNTK performance')
plt.ylabel('CNN performance')
plt.title('Table 1 of Arora et. al., Correlation betwen NTK and CNN?')
#plt.text(64,87,'T = 0.511',{'size':16,'color':'tab:green'})
plt.ylim(55,90)
```

```python
CNTK2 = [[92.15,92.80,93.10,93.05,91.95],[71.05,94.50,95.25,95.40,95.70]]
RF = [[87.25,87.78,88.73,87.80,85.35],[51.10,52.85,53.98,56.55,90.65]]
```

```python
print(kendalltau(RF[0],CNTK2[0]))
print(kendalltau(RF[1],CNTK2[1]))
```

```python
plt.plot(RF[0],CNTK2[0],'k.')
plt.plot(RF[1],CNTK2[1],'.',color='tab:blue')
plt.xlabel('RF performance')
plt.ylabel('CNTK performance')
plt.title('Table 2 of Arora et. al., Correlation between finite NTK and CNTK?')
#plt.text(50,55,'T = 0.999',{'size':16,'color':'tab:green'})
plt.ylim(50,100)
```
